## Typically, you would download the data from Kaggle, but here I am just pulling it from someone else's Github.

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
titanic.head(10)
```

### Create a numerical (binary) column based on the column 'alive'

```python
# create a column called column_alive which is = 1 if alive is yes and 0 if alive is no

titanic['alive_yes'] = np.where(titanic['alive'] == 'yes', 1, 0)
titanic['Female'] = np.where(titanic['sex'] == 'female', 1, 0)
titanic['alone_yes'] = np.where(titanic['alone'] == True, 1, 0)
titanic['column_alive'] = np.where(titanic['alive'] == 'yes', 1, 0)
titanic.head(10)
```

```python
print('The names of the columns are:', titanic.columns)
print('The Index', titanic.index)
```

```python
# Type of each column
titanic.dtypes
```

```python
#Average age
titanic['age'].mean()
```

```python
#Total Revenue
titanic['fare'].sum()
```

```python
#Survivors and % of survivers
print(titanic['survived'].sum())
print(titanic['survived'].mean())
```

```python
# Total men, total men survived, total men per pclass, total men per p class average
titanic['sex'].value_counts()
titanic[titanic['sex'] == 'male']['survived'].sum()
titanic[titanic['sex'] == 'male'].groupby('pclass')['survived'].sum()
titanic[titanic['sex'] == 'male'].groupby('pclass')['survived'].mean()

titanic[titanic['sex'] == 'female']['survived'].sum()
titanic[titanic['sex'] == 'female'].groupby('pclass')['survived'].sum()
titanic[titanic['sex'] == 'female'].groupby('pclass')['survived'].mean()
```

```python
# Total children, total boys
titanic[titanic['age'] < 18]['survived'].sum()
print(titanic[titanic['age'] < 18]['survived'].sum())
titanic[titanic['age'] < 18]['survived'].mean()
print(titanic[titanic['age'] < 18]['survived'].mean())
titanic[(titanic['age'] < 18) & (titanic['sex'] == 'male')]['survived'].sum()
```

```python
titanic.groupby('pclass')['survived'].mean()

```

```python
titanic[titanic['embark_town'] == 'Southampton'].shape[0]
titanic[titanic['embark_town'] == 'Southampton']['fare'].sum()
titanic[titanic['embark_town'] == 'Southampton']['fare'].mean()
# Change female to male to see male sum
titanic[titanic['embark_town'] == 'Southampton']['sex'].value_counts()
titanic[titanic['embark_town'] == 'Southampton'][titanic['sex'] == 'female']['survived'].sum()
titanic[titanic['embark_town'] == 'Southampton'][titanic['sex'] == 'female'].groupby('pclass')['survived'].sum()

titanic[titanic['embark_town'] == 'Southampton']['age'].value_counts()
titanic[titanic['embark_town'] == 'Southampton'][titanic['age'] < 18]['survived'].sum()


```

```python
print(titanic[titanic['alone'] == True]['survived'].sum())
print(titanic[titanic['alone'] == True]['survived'].mean())
```

```python
#Oldest man and woman and features
oldest_man = titanic[titanic['sex'] == 'male'].sort_values('age', ascending = False).head(1)[['age', 'pclass', 'embark_town', 'survived']]
oldest_woman = titanic[titanic['sex'] == 'female'].sort_values('age', ascending = False).head(1)[['age', 'pclass', 'embark_town', 'survived']]
print(oldest_man)
print(oldest_woman)
```
